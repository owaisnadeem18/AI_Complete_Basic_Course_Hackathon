# Accessing the Data Now

```python
SeriesExample = pd.Series(["Ahmed" , "Ali" , "Faisal" , "Akbar" , "Saleem" , "Daniyal"] , index=[ 1, 2 ,3 , 4 ,5 ,6 ])

SeriesExample
```

```python

student = {
            "Roll_Numbers" : [20 ,34 ,78 ,54 , 82 ,13],
           'Name' : ["Ali" , "Owais" , "Hamza" , "Fakhar" , "Saim" , "Hammad"],
           'Mode' : ["Online " , "Onsite" , "Online","Online " , "Onsite" , "Online"],
           'courses' : [ "PIAIC" , "SMIT" , "Both" , "PIAIC" , "SMIT" , "Both"],
           'Gender' : ['Male' , 'Male' , 'Male','Male' , 'Male' , 'Male'],
           'Age' : [45 , 23 , 12 , 89 ,65 ,34],
           'Status' : [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]

}
```

```python
student
```

```python
data = pd.DataFrame(student)
data
```

```python
data['Status'] = ["Passed" if age<50 else "Fail" for age in data["Age"]]

data
```

Add a new column in our DataFrame through pandas

```python
data["City"] = data["Mode"].apply(lambda x: 'Karachi ' if x == "Onsite" else "Outside Khi")

data
```

```python
def Age_Func(Age):
  if Age<20:
    return ("teenager")
  elif Age<40:
    return ("Young")
  elif Age<60:
    return ("Experience Man")
  elif Age>60:
    return ("retired")

Age_Func(134)
```

```python
data["Age Factor"] = data["Age"].apply(Age_Func)
data

# data.drop("Status_Of_Age" , axis =1 )
```
